```python
import pandas as pd  
import numpy as np
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
driver=webdriver.Chrome(service=Service(r"C:\Users\Shubham\Downloads\chromedriver-win64\chromedriver.exe"))

driver.get('https://housing.com/in/buy/searches/AB1AC2CuH2Pskwz0ocdh7q42r5?gad_source=1&gclid=Cj0KCQjwgJyyBhCGARIsAK8LVLNtTUpVZvAdRV86VjIMQrws_oazHDRY87eQRpwVssGsgYlyfsOoylgaAnv2EALw_wcB')

time.sleep(2)

original_window = driver.current_window_handle

flat_type = []
price = []
address = []
build_up_area = []
avg_price_sqft = []
age_of_property = []
possesion_status = []
which_floor = []
facing = []
furnicing = []
project_name = []
brokrage = []
carpet_area = []
bedrooms = []
bathrooms = []
parking = []
balcony = []
about_property = []

# loop to click on element
for i in range(2,5000):
    try:
        # path to get the button to click
        xpath = f'//*[@id="srp-{i}"]/div[1]/div[2]/div[1]/div[1]/div[1]/div'
        element = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, xpath)))
        # click on element
        element.click()
        time.sleep(2)
        # switch to new window
        driver.switch_to.window(driver.window_handles[1])
        # get the html code
        page = driver.page_source
        soup = BeautifulSoup(page,'lxml')
        # closing the new window
        driver.close()
        # switch to original window
        driver.switch_to.window(original_window)
        time.sleep(1)
        try:
            # fetching type of flat
            flat_type.append(soup.find('div',class_='css-1hidc9c').text)
        except:
             flat_type.append(np.nan) 

        try:
            # fetching price
            price.append(soup.find('span',class_='css-gg3jj9').text)
        except:
            price.append(np.nan)

        try:
            # address
            address.append(soup.find('div',class_='css-1ty5xzi').text)
        except:
            address.append(np.nan)
        
        try:
            build_up_area.append(soup.find_all('div',class_='css-1k19e3')[0].text)
        except:
            build_up_area.append(np.nan)

        try:
            avg_price_sqft.append(soup.find_all('div',class_='css-1k19e3')[1].text)
        except:
            avg_price_sqft.append(np.nan)    

        try:
            age_of_property.append(soup.find_all('div',class_='css-1k19e3')[2].text)
        except:
            age_of_property.append(np.nan)   

        try:
            possesion_status.append(soup.find_all('div',class_='css-1k19e3')[3].text)
        except:
            possesion_status.append(np.nan)

        try:
            which_floor.append(soup.find_all('div',class_='css-c2zxhw')[4].text)
        except:
            which_floor.append(np.nan)

        try:
            facing.append(soup.find_all('div',class_='css-1k19e3')[5].text)    
        except:
            facing.append(np.nan)     

        try:
            furnicing.append(soup.find_all('div',class_='css-1k19e3')[6].text)    
        except:
            furnicing.append(np.nan)  

        try:
            project_name.append(soup.find_all('td',class_='_9s1txw _w49jpi _ar1bp4 T_55d05cd6')[0].text)
        except:
             project_name.append(np.nan)
        
        try:
            brokrage.append(soup.find_all('td',class_='_9s1txw _w49jpi _ar1bp4 T_55d05cd6')[1].text)
        except:
            brokrage.append(np.nan)   

        try:
            carpet_area.append(soup.find_all('td',class_='_9s1txw _w49jpi _ar1bp4 T_55d05cd6')[3].text)
        except:
            carpet_area.append(np.nan)                          
        
        try:
            bedrooms.append(soup.find_all('td',class_='_9s1txw _w49jpi _ar1bp4 T_55d05cd6')[4].text)
        except:
            bedrooms.append(np.nan)

        try:
            bathrooms.append(soup.find_all('td',class_='_9s1txw _w49jpi _ar1bp4 T_55d05cd6')[5].text)
        except:
            bathrooms.append(np.nan)

        try:
            balcony.append(soup.find_all('td',class_='_9s1txw _w49jpi _ar1bp4 T_55d05cd6')[6].text)
        except:
            balcony.append(np.nan)     

        try:
            parking.append(soup.find('td',class_='_9s1ule T_496aca80').text)
        except:
            parking.append(np.nan)  

        try:
            about_property.append(soup.find('div',class_='T_5255c66f _lorj18uv _v1ivgktf _g31tcg _7l9wsg _h3f6fq T_a3fd8ac3 _1ln11vji').text)
        except:
            about_property.append(np.nan)         
    except:
            continue
```

```python
df = pd.DataFrame({
    'flat_type':flat_type,
    'price':price,
    'address':address,
    'buildup_area':build_up_area,
    'avg_price_sqft':avg_price_sqft,
    'age_of_property':age_of_property,
    'possesion_status':possesion_status,
    'which_floor':which_floor,
    'facing':facing,
    'furnicing':furnicing,
    'project_name':project_name,
    'brokrage':brokrage,
    'carpet_area':carpet_area,
    'bedrooms':bedrooms,
    'bathrooms':bathrooms,
    'balcony':balcony,
    'parking':parking,
    'about_property':about_property
})
```

```python
df.to_csv("real_estate.csv",index=False)
```
